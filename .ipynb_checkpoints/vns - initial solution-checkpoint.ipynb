```python
import random 
import time
```

```python
def initial_solution(A, f):
    start = time.time()
    
    len_graph = len(A)

    mark = {x: False for x in range(len_graph)} 

    k = random.randint(0, len_graph - 1) 
    
    mark[k] = True 
    
    q = [] 
    
    q.append(k) 

    ql = 1 
    
    f = {x: -1 for x in range(len_graph)}
    f[k] = 0 

    l = 0 
    
    while l < len_graph: 
        
        r = 0 
        s = [] 

        for i1 in range(ql): 
            i = q[i1] 

            for j1 in range(len(A[i])):
                j = A[i][j1]
                
                if not mark[j]: 
                    s.append(j) 
                    r += 1 
                    mark[j] = True 

        if ql == 1:
            j_ = 0 
        else:
            j_ = random.randint(0, ql - 1) 
    
        for jl in range(ql):
            j = q[j_] 
            l += 1 
            f[j] = l 
            j_ += 1 
            
            if (j_ >= ql): 
                j_ = 0
        q = s
        ql = r 
        
    end = time.time()
         
    return f, (end - start)
```

```python
def create_adjacency_list(nome):
    a = open(nome + ".mtx", 'r')
    start = False
    
    for line in a: 
        if line[0] != "%":
            data = list(map(float, line.split()))
            if not start:
                adjList = [ [] for i in range (int(data[0]))]
                start = True
            else:
                adjList[int(data[0] - 1)].append(int(data[1] - 1))
                if data[0] != data[1]:
                    adjList[int(data[1] - 1)].append(int(data[0] - 1))
                
    return adjList
```

```python
def labelling_adjacency_list(len_mat, labels):
    labeled_adjacency_list = [ [] for i in range(len_mat)]
    for key, value in labels.items():
        labeled_adjacency_list[int(key)].append(int(value))
    return labeled_adjacency_list
```

```python
def calculate_bandwidth(A):
        index_vertex = 0
        largest_bandwidth = -9999999999
        
        for vertex in A: 
            bandwidth_vertex = index_vertex - (min(vertex))
            index_vertex += 1
            
            if bandwidth_vertex > largest_bandwidth:
                largest_bandwidth = bandwidth_vertex
                
        return largest_bandwidth
```
