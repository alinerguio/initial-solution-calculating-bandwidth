```python
import random 
from scipy.io import mmread
import networkx as nx
import numpy as np
import time
```

```python
def initial_solution(A, f):
    start = time.time()
    
    len_graph = len(A)

    mark = {x: False for x in range(len_graph)} # lista que marca um vértice como visitado

    k = random.randint(0, len_graph - 1) # posição aleatória para indicar o vértice a iniciar
    
    mark[k] = True # como foi primeiro vistado, recebe true

    q = [] # sequência de vértices visitados 
    
    q.append(k) # primeiro vértice a ser visitado

    ql = 1 # numero de vertices a serem explorados na proxima iteracao 
    
    f = {x: -1 for x in range(len_graph)} # guarda os labels de cada vértice (de acordo com sua entrada)
    f[k] = 0 

    l = 0 # primeiro label (atribuido ao vértice k)
    
    while l < len_graph: # enquanto todos os labels não forem atribuídos
        
        r = 0 # marcados adjacentes a A[i]
        s = [] # sequencia dos vértices visitados (auxiliar)

        for i1 in range(ql): # para cada ligação do vértice atual
            i = q[i1] # para cada vértice visitado 

            for j1 in range(len(A[i])): # para cada ligação do vértice
                j = A[i][j1] # vértice adjacente a i
                
                if not mark[j]: # se não foi visitado ainda
                    s.append(j) # sequencia dos vértices visitados
                    r += 1 # numero de marcados adjacentes a A[i]
                    mark[j] = True # marcado como visitado
        
        #alteração feita porque não pode ser 0

        if ql == 1:
            j_ = 0 # primeira posicao é 0
        else:
            j_ = random.randint(0, ql - 1) # aleatório na lista de ligações (posições começam em 0)
    
        for jl in range(ql):
            j = q[j_] # pega o valor da posição (PODE DAR -1) - porque é inicia
            l += 1 # muda o label
            f[j] = l # atribui o label se for válida a posição
            j_ += 1 # ver mais posições dentro do q dentro do for
            
            if (j_ >= ql): # se for o último possível ou maior (ja que se for ultimo a escolhido)
                j_ = 0
        q = s
        ql = r # número de vértices a serem explorados

    end = time.time()
         
    return f, (end - start)
```

```python
def create_adjacency_list(nome):
    a = open(nome + ".mtx", 'r')
    start = False
    
    for line in a: 
        if line[0] != "%":
            data = list(map(float, line.split()))
            if not start:
                adjList = [ [] for i in range (int(data[0]))]
                start = True
            else:
                adjList[int(data[0] - 1)].append(int(data[1] - 1))
                if data[0] != data[1]:
                    adjList[int(data[1] - 1)].append(int(data[0] - 1))
                
    return adjList
```

```python
def is_symmetric(a):
    if a.shape[0] != a.shape[1]:
        raise Exception("Matriz não é quadrada!")
    return np.allclose (a.A, (a.transpose()).A) 
```

```python
def labelling_adjacency_list(A, labels):
    labeled_adjacency_list = []
    for key, value in labels.items():
        labeled_adjacency_list.append(A[value - 1])
    return labeled_adjacency_list
```

```python
def calculate_bandwidth(A):
        index_vertex = 0
        largest_bandwidth = -9999999999
        
        for vertex in A: 
            bandwidth_vertex = index_vertex - (min(vertex))
            index_vertex += 1
            
            if bandwidth_vertex > largest_bandwidth:
                largest_bandwidth = bandwidth_vertex
                
        return largest_bandwidth
```
